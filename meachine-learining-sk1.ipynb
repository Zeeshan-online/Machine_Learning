```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression, LogisticRegression


df = sns.load_dataset('tips')
# df.head()
X = df [['total_bill', 'size']]

y = df ['tip']

model = LinearRegression()

model.fit(X, y)

model.predict([[35.527, 2]])
```

```python
df = sns.load_dataset('tips')
# encode smoker column in numaric

df['smoker_num'] = df['smoker'].map({'Yes': 1, 'No': 0})

# df.head()

X = df [['total_bill', 'size', 'smoker_num']]

y = df ['tip']

model = LinearRegression()

model.fit(X, y)

print(model.predict([[35.527, 5, 1]]))
print(model.predict([[35.527, 5, 0]]))
```

```python
df = sns.load_dataset('tips')
# encode smoker column in numaric


X = df [['total_bill', 'size', 'tip']]

y = df ['smoker']

model = LogisticRegression()

model.fit(X, y)

# predict
model.predict([[50.527, 2, 5.67]])
```

## Decision Trees classifier

```python
from sklearn.tree import DecisionTreeClassifier

df = sns.load_dataset('tips')
# encode smoker column in numaric


X = df [['total_bill', 'size', 'tip']]

y = df ['smoker']

model = DecisionTreeClassifier()

model.fit(X, y)

# predict
model.predict([[50.527, 2, 5.67]])
```

# Regression Metrics

```python
df = sns.load_dataset('tips')
# encode smoker column in numaric


X = df [['total_bill', 'size', ]]

y = df ['tip']

# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

# call the model

model = LinearRegression()

# fit the model

model.fit(X_train, y_train)

# metric to evalutate the model

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error,mean_absolute_percentage_error
y_pred = model.predict(X_test)

print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"R2_Score Error: {r2_score(y_test, y_pred)}")
print(f"Mean Absoulate Error: {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Absoulate Percentage Error: {mean_absolute_percentage_error(y_test, y_pred)}")

# root mean squared error

print(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(y_test, y_pred))}")
```
